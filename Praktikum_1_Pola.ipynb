```python
import numpy as sp
traffic = sp.genfromtxt("web_traffic.tsv",delimiter='\t')
print(traffic[:10])
print(traffic.shape)
x = traffic[:,0]
y = traffic[:,1]
x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.title("Web traffic last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
plt.xticks([w*7*24 for w in range(10)],['week %i' %w for w in
range(10)])
plt.autoscale(tight=True)
plt.grid()
```

```python
resized = cv2.resize(babon, (150, 150))  # Mengubah ukuran ke 50x50 piksel
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
cropped = babon[50:200, 100:300]  # Memotong gambar dari koordinat (y1:y2, x1:x2)
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread("ipb.png")
flipped = cv2.flip(image, 1)  # 1 untuk flip horizontal, 0 untuk flip vertikal, -1 untuk flip keduanya
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread("ipb.png")
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

# Rotasi 45 derajat
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread("ipb.png")
blurred = cv2.GaussianBlur(image, (15, 15), 0)  # Kernel 15x15 untuk efek blur
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image = cv2.imread("ipb.png")
edges = cv2.Canny(image, 100, 200)  # Threshold 100-200
plt.imshow(edges, cmap="gray")
plt.show()
```
